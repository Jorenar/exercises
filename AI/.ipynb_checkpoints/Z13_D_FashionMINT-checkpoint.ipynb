```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train_full.shape
```

Each pixel intensity is represented as a byte (0 to 255):

```python
X_train_full.dtype
```

* Podzielmy pełny zestaw treningowy na zestaw walidacyjny i zestaw treningowy. 
* Kolor pikseli skalujemy również do zakresu 0-1 i przekształcamy na zmiennoprzecinkowe, dzieląc przez 255.

```python
X_valid, X_train = X_train_full[:5000] / 255., X_train_full[5000:] / 255.
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
X_test = X_test / 255.
```

```python
plt.imshow(X_train[0], cmap="binary")
plt.axis('off')
plt.show()
```

```python
y_train
```

Będziemy używać klas numerycznych i nazw.

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

class_names[y_train[0]]
```

Możemy narysować więcej przykładów treningowych.

```python
n_rows = 4
n_cols = 10
plt.figure(figsize=(n_cols * 1.2, n_rows * 1.2))
for row in range(n_rows):
    for col in range(n_cols):
        index = n_cols * row + col
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(X_train[index], cmap="binary", interpolation="nearest")
        plt.axis('off')
        plt.title(class_names[y_train[index]], fontsize=12)
plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
```

# Zad.
Porównaj wyniki architektury fully connected (najlepszej jaką uda ci się zbudować) i architektury konwolucyjnej (najlepszej jaką uda ci się zbudować).


Wizualizuj wyniki:

* porównaj krzywe uczenia


```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation
from keras.layers import Flatten
from keras.callbacks import History
from keras.layers.convolutional import Conv2D, MaxPooling2D, AveragePooling2D
keras.backend.clear_session()

full_connected_history = History()

full_connected_model = Sequential()
full_connected_model.add(Flatten(input_shape=[28, 28]))
full_connected_model.add(Dense(300, activation="relu"))
full_connected_model.add(Dense(100, activation="relu"))
full_connected_model.add(Dense(10, activation="softmax"))
full_connected_model.summary()

full_connected_model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])

full_connected_model.fit(X_train, y_train, epochs=30,
                    validation_data=(X_valid, y_valid), 
                    callbacks=[full_connected_history])
```

```python
keras.backend.clear_session()

cnn_history = History()

cnn_model = Sequential()
cnn_model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
cnn_model.add(MaxPooling2D((2, 2)))
cnn_model.add(Flatten())
cnn_model.add(Dense(10,activation="softmax"))
cnn_model.compile(optimizer="sgd", loss='sparse_categorical_crossentropy', metrics=['accuracy'])

cnn_model.fit(X_train, y_train, epochs=30,
                    validation_data=(X_valid, y_valid), 
                    callbacks=[cnn_history])
```

```python
cnn_model.summary()
```

```python
plt.plot(full_connected_history.history['accuracy'], label = "tarina fully con")
plt.plot(full_connected_history.history['val_accuracy'], label = "test fully con")

plt.plot(cnn_history.history['accuracy'], label = "tarina conv")
plt.plot(cnn_history.history['val_accuracy'], label = "test conv")

plt.legend()
plt.show()
```
